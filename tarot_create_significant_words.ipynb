```python
import pandas as pd

t = pd.read_csv("tarot_with_text.csv")

t["merged"] = t["h"] + " \n " + t["i"]
t["merged"]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(stop_words='english', token_pattern=r'[a-zA-Z]+', min_df=0, max_df=0.017, ngram_range=(1,1))
X = vectorizer.fit_transform(t["merged"])
frequency_words = pd.DataFrame(X.toarray(),columns=vectorizer.get_feature_names())
frequency_words
```

```python

cards_words = []
transposed = frequency_words.T
for card_index in transposed:
    card_words = []
    for word, frequency in zip(transposed.index, transposed[card_index]):
        if 0 < frequency < 0.7:
            card_words += [word]
    cards_words += [" ".join(card_words)]

t["main_words"] = cards_words
#
# t["main_words_count"] = [len(words) for words in t["main_words"]]
# t.hist(column="main_words_count")
```

```python
df_to_append = pd.read_csv("tarot.csv")
df_to_append["words"] = t["main_words"]
df_to_append.to_csv("tarot.csv")
df_to_append
```
